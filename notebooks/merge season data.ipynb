```python
import os
import glob
import pandas as pd
```

```python
# Used same code snippet for each season
# Downloaded monthly csv files, renamed it to data_year-month 
# Followed identical procedure for each subsequent season
# 2017-2018 season
all_files = glob.glob("data_*.csv")
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged_2017_2018   = pd.concat(df_from_each_file, ignore_index=True)
df_merged_2017_2018.to_csv("2017_2018_merged.csv")
```
