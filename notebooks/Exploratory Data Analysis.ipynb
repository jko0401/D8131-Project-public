```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diff_in_wins = pd.DataFrame(columns=['Season', 'Wins - Predicted Wins'])
```

```python
preseason_1718 = pd.read_csv("preseason odds 2017-2018.csv")
preseason_1718 = preseason_1718.loc[:, preseason_1718.columns!='Unnamed: 2']
preseason = preseason_1718.groupby('Team').mean()["Odds"].sort_values(ascending=True).plot.bar(title = "Preseason Odds in 2017-2018", ylabel = "Odds").get_figure()
plt.show()
preseason.savefig("preseason1718.png")
```

```python
preseason_1718.to_csv("1718odds.csv")
```

```python
preseason_1718["Wins"] = [int(preseason_1718["Result"][i].split("-")[0]) for i in range(len(preseason_1718))] 
preseason_1718.head()
```

```python
preseason_1718["Wins - Predicted Wins"] = [preseason_1718["Wins"][i] - preseason_1718["W-L O/U"][i] for i in range(len(preseason_1718))]
preseason_1718.head()
```

```python
preseason = preseason_1718.groupby('Team').mean()["Wins - Predicted Wins"].sort_values(ascending=True).plot.bar(title = "Difference in Expected Wins for 2017-2018 Season", ylabel = "Wins - Predicted Wins").get_figure()
plt.show()
preseason.savefig("preseasondiff1718.png")
```

```python
preseason_1718.sort_values(by=['Wins - Predicted Wins']).to_csv("1718diffinwins.csv")
```

```python
diff_in_wins.loc[len(diff_in_wins.index)] = ["2017-2018 Season", preseason_1718["Wins - Predicted Wins"].mean()] 
diff_in_wins
```

```python
data_1718 = pd.read_csv("2017_2018_merged.csv")
data_1718 = data_1718.rename(columns={"Unnamed: 0": "Gameid", "PTS": "Visitor/Neutral Pts", "PTS.1": "Home/Neutral Pts", "Unnamed: 6": "Boxscore", "Unnamed: 7": "Overtime"})
data_1718.head()
```

```python
# number of games in 2017-2018
len(data_1718)
```

```python
visitor = data_1718.groupby('Visitor/Neutral').mean()["Visitor/Neutral Pts"].sort_values(ascending=False).plot.bar(title = "Visitor/Neutral Avg Pts in 2017-2018 Season", ylabel = "Pts").get_figure()
plt.show()
visitor.savefig("visitor1718.png")
```

```python
visitor = data_1718.groupby('Visitor/Neutral').mean()["Home/Neutral Pts"].sort_values(ascending=True).plot.bar(title = "Visitor/Neutral Avg Pts Allowed in 2017-2018 Season", ylabel = "Pts").get_figure()
plt.show()
visitor.savefig("visitorDef1718.png")
```

```python
home = data_1718.groupby('Home/Neutral').mean()["Home/Neutral Pts"].sort_values(ascending=False).plot.bar(title = "Home/Neutral Avg Pts in 2017-2018 Season", ylabel = "Pts").get_figure()
plt.show()
home.savefig("home1718.png")
```

```python
home = data_1718.groupby('Home/Neutral').mean()["Visitor/Neutral Pts"].sort_values(ascending=True).plot.bar(title = "Home/Neutral Avg Pts Allowed in 2017-2018 Season", ylabel = "Pts").get_figure()
plt.show()
home.savefig("homedef1718.png")
```

```python
teams = data_1718["Visitor/Neutral"].unique()
avg_pts = pd.DataFrame(columns=['Team', 'Pts'])
for name in teams:
    values = data_1718[(data_1718 == name).any(axis=1)]
    away = np.where(values["Visitor/Neutral"] == name, values["Visitor/Neutral Pts"], "")
    away = away[away != '']
    away = away.astype(np.float)
    home = np.where(values["Home/Neutral"] == name, values["Home/Neutral Pts"], "")
    home = home[home != '']
    home = home.astype(np.float)
    total = np.append(away,home)
    total = np.mean(total)
    avg_pts.loc[len(avg_pts.index)] = [name, total] 
    
team_avg = avg_pts.groupby('Team').mean()["Pts"].sort_values(ascending=False).plot.bar(title = "Avg Pts in 2017-2018 Season", ylabel = "Pts").get_figure()
plt.show()
team_avg.savefig("1718.png")
```

```python
avg_pts.groupby('Team').mean()["Pts"].sort_values(ascending=False).to_csv("1718points.csv")
```
