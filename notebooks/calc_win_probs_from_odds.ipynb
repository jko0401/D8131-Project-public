```python
import pandas as pd
import os
import glob
cwd = os.getcwd()
path = cwd+'/odds_parsed_csv/*.csv'
```

```python
df = pd.read_csv(cwd+'/odds_parsed_csv/2017-18_odds_parsed.csv')
df = df.drop(columns=['Unnamed: 0'], axis=1)
df.head()
```

```python
def calc_prob(ml):
    if ml < 0: p = -ml/(-ml+100)
    else: p = 100/(ml+100)
    return p

def normalize(vp, hp):
    return vp/(vp+hp), hp/(vp+hp)
```

```python
test = normalize(calc_prob(166), calc_prob(-195))
test
```

```python
df['v_win_prob'] = df[['ml_v','ml_h']].apply(lambda x: normalize(calc_prob(x['ml_v']),calc_prob(x['ml_h']))[0],axis=1)
df['h_win_prob'] = df[['ml_v','ml_h']].apply(lambda x: normalize(calc_prob(x['ml_v']),calc_prob(x['ml_h']))[1],axis=1)
```

```python
path = cwd+'/odds_parsed_csv/*.csv'
for fname in glob.glob(path):
    df = pd.read_csv(fname)
    df = df.drop(columns=['Unnamed: 0'], axis=1)
    df['v_win_prob'] = df[['ml_v','ml_h']].apply(lambda x: normalize(calc_prob(x['ml_v']),calc_prob(x['ml_h']))[0],axis=1)
    df['h_win_prob'] = df[['ml_v','ml_h']].apply(lambda x: normalize(calc_prob(x['ml_v']),calc_prob(x['ml_h']))[1],axis=1)
    df.to_csv(fname.strip('.csv')+'probs.csv')
```
